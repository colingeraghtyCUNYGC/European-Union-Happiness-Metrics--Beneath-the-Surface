```python
eugdp = pd.read_excel('EUGDP.xlsx')
eudi = pd.read_excel('EUDI.xlsx')
eugdp_2021 = eugdp[['Country Name', '2021 [YR2021]']]
merged_data = pd.merge(eugdp_2021, eudi, left_on='Country Name', right_on='TIME', how='inner')
merged_data = merged_data.drop(columns='TIME')
print(merged_data)



```

```python
import statsmodels.api as sm

# Set the dependent variable (GDP) and independent variable (disposable income)
Y = merged_data['2021 [YR2021]']
X = merged_data['2021']
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(Y, X, missing='drop').fit()

# Display the summary
print(model.summary())
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Visualizing the linear relationship
p = sb.lmplot(x='2021', y='2021 [YR2021]', data=merged_data)
p.fig.set_figwidth(20)
p.fig.set_figheight(12)
plt.xlabel("Adjusted Gross Disposable Income of Households per Capita")
plt.ylabel("GDP for 2021")
plt.title("Relationship between Disposable Income and GDP for EU Countries in 2021")
plt.show()

```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Set up the plot
plt.figure(figsize=(20, 12))
ax = plt.gca()

# Plot the regression line and scatter plot
sb.regplot(x='2021', y='2021 [YR2021]', data=merged_data, ax=ax)

# Calculate residuals for each point
merged_data['residual'] = merged_data['2021 [YR2021]'] - (merged_data['2021'] * model.params['2021'] + model.params['const'])

# Annotate each point with its country name
for i, row in merged_data.iterrows():
    # If the absolute residual is large (e.g., greater than a threshold), highlight the label
    if abs(row['residual']) > (merged_data['residual'].std() * 1.5): # 1.5 times the standard deviation as a threshold
        ax.annotate(row['Country Name'], (row['2021'], row['2021 [YR2021]']), color='red')
    else:
        ax.annotate(row['Country Name'], (row['2021'], row['2021 [YR2021]']))

# Set labels and title
plt.xlabel("Adjusted Gross Disposable Income of Households per Capita")
plt.ylabel("GDP for 2021")
plt.title("Relationship between Disposable Income and GDP for EU Countries in 2021")
plt.show()



```

```python
import seaborn as sb
import matplotlib.pyplot as plt

#  percentage of GDP that is disposable income for each country
merged_data['Percentage'] = (merged_data['2021'] / merged_data['2021 [YR2021]']) * 100

#  plot
plt.figure(figsize=(20, 12))
ax = plt.gca()

# regression line and scatter plot
sb.regplot(x='2021', y='2021 [YR2021]', data=merged_data, ax=ax, scatter_kws={'s':100})  # 's':100 increases the size of scatter points

# Annotate each point with its percentage value
for i, row in merged_data.iterrows():
    ax.annotate(f"{row['Percentage']:.2f}%", (row['2021'], row['2021 [YR2021]']),
                textcoords="offset points", xytext=(0,10), ha='center')

# Set labels and title
plt.xlabel("Adjusted Gross Disposable Income of Households per Capita")
plt.ylabel("GDP for 2021")
plt.title("Relationship between Disposable Income and GDP for EU Countries in 2021 with Disposable Income as a Percentage of GDP")
plt.show()
```
